```python
import graspy

import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from graspy.plot import heatmap
import heapq
np.random.seed(88889999)

%matplotlib inline
```

```python
def draw_MMvectors(alpha, n):
    MMvectors = np.random.default_rng().dirichlet(alpha, n)
    return MMvectors
def draw_MMindicator(MMVector):
    MMindicator = np.random.multinomial(1, MMVector, size = 1)
    return MMindicator
def sample_edge(MMindicator_pq, MMindicator_qp, rho):
        edge_pq = np.random.binomial(1, (1.0-rho)*(MMindicator_pq @ B @ MMindicator_qp.T))
        #edge_pq = np.random.binomial(1, (((1.0-rho)*MMindicator_pq) @ B @ MMindicator_qp.T))
        #edge_qp = np.random.binomial(1, ((MMindicator_qp) @ B @ MMindicator_pq.T))
        edge_qp = np.random.binomial(1, (1.0-rho)*(MMindicator_qp @ B @ MMindicator_pq.T))
        return (edge_pq, edge_qp)
```

```python
n = 100
k = 4
alphas = [0.05, 0.1, 0.25]
for val in alphas:

    alpha = val*np.ones(k)
    rho = 0.0

    B = [[0.2, 0.7, 0.2, 0.2],
        [ 0.2, 0.7, 0.7, 0.7],
        [ 0.7, 0.2, 0.2, 0.7],
        [ 0.7, 0.2, 0.2, 0.7]]


    MMvectors = draw_MMvectors(alpha = alpha, n= n)

    indeces_same_groups = [[index for index,val in enumerate(MMvector) if val == np.max(MMvector)] for MMvector in MMvectors]

    smallest_k = heapq.nsmallest(n,((m, g) for g, m in enumerate(indeces_same_groups)))
    _, index_smallest =  zip(*smallest_k)

    idx = np.array(index_smallest)

    MMvectors = MMvectors[idx]

    A = np.zeros((n,n))

    for i in range(n):
        for j in range(n):
            if(i >j):
                Z_pq = draw_MMindicator(MMvectors[i])
                Z_qp = draw_MMindicator(MMvectors[j])
                #print(Z_pq)
                #print(Z_qp)
                edges = sample_edge(MMindicator_pq = Z_pq, MMindicator_qp = Z_qp, rho = rho)
                A[i,j] = edges[0]
                A[j,i] = edges[1]
    print(A)

    heatmap(A)
```

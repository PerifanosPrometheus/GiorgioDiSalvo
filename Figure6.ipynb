```python
import numpy as np

from graspologic.simulations import MMsbm,sbm
from graspologic.plot import heatmap

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 100
k = 4
alphas = [0.05, 0.1, 0.25]
for val in alphas:
    
    rng = np.random.default_rng(1234)

    alpha = val*np.ones(k)
    rho = 0.0

    B = [[0.2, 0.7, 0.2, 0.2],
        [ 0.2, 0.7, 0.7, 0.7],
        [ 0.7, 0.2, 0.2, 0.7],
        [ 0.7, 0.2, 0.2, 0.7]]
    G = MMsbm(n, p=B, alpha = alpha, rng = rng, directed = True)
    heatmap(G)
```
